```python
import pandas as pd

df = pd.read_csv('good_loans.csv')
```

```python
df.head()
```

```python
# make a dummy array so we can insert a new column with the return from the loans using compound interest
dummy_array=[]
counter=0

# build a dummy array of the same length to fill our new columns
for x in df['loan_amnt']:
    dummy_array.append(counter)
    counter=counter+1
```

```python
# Insert a new column for the total returns
df.insert(7,"returns",dummy_array)
```

```python
# Insert a new column for the multiplying the installments by the term
df.insert(8,"installments_and_terms", dummy_array)
```

```python
# Insert a new column for the annualized returns, which is the total returns - the initial loan amount
df.insert(7,"annualized_returns",dummy_array)
```

```python
#function to get the returns by taking the payment installments and multiplying it by the term length
def get_installment_and_terms(row):
    return int(row['term'][:3])*row['installment']

#function to get the total returns using the compound interest formula
def get_returns(row):
    return round(row['loan_amnt']*((1+((row['int_rate']/100)/12))**(int(row['term'][:3])/2)),2)

#function to get the annualized returns
def get_annualized_returns(row):
    return row['returns']-row['loan_amnt']
```

```python
# for row in df.iterrows():
#     row[1]['returns'] = row[1]['returns'] -1
#     print(row[1]['returns'])


# populate the new columns with the correct values
df['returns'] = df.apply (lambda row: get_returns(row), axis=1)
df['installments_and_terms'] = df.apply (lambda row: get_installment_and_terms(row), axis=1)
df['annualized_returns'] = df.apply (lambda row: get_annualized_returns(row), axis=1)
```
